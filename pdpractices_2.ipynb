```python
import pandas as pd
import os
```

```python
df = pd.read_csv("./pdpracs/Sales_April_2019.csv")
files = [file for file in os.listdir('./pdpracs')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./pdpracs/"+file)
    all_months_data = pd.concat([all_months_data,df])

all_months_data.head()
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()
```

```python
all_data = all_data.dropna(how = 'all')
all_data.head()
```

```python
all_data = all_data[all_data["Order Date"].str[0:2] != 'Or']
all_data.head()
```

```python
all_data['Month'] =all_data['Order Date'].str[0:2] 
all_data['Month'] =all_data['Month'].astype('int32')
all_data.head()
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] =all_data['Quantity Ordered']*all_data['Price Each']
 
```

```python
all_data.head()
```

```python
results = all_data.groupby('Month').sum()

results.head()
```

```python
import matplotlib.pyplot as plt
months = range(1,10)
plt.bar(months,results["Sales"])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month Number')
```

2.What city has the highest number of sales?

```python
def get_city(address):
    return address.split(',')[1]
def get_states(address):
    return address.split(',')[2].split()[0]
```

```python
all_data["City"] = all_data['Purchase Address'].apply(lambda x : f"{get_city(x)}({get_states(x)})")
all_data.head(20)
```

```python
result2 = all_data.groupby('City').sum()
result2
```

```python
import matplotlib.pyplot as plt
Cities = [city for city,df in all_data.groupby('City')]
plt.bar(Cities,result2["Sales"])
plt.xticks(Cities,rotation = 'vertical',fontsize = 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')
```
